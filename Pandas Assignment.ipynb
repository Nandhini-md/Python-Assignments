```python
#Q10. How do you merge two Pandas DataFrames on a specific column?
ndf1=pd.DataFrame({'Employee Name':['John', 'Zeke', 'Max','Jenny'],
                    'Employee ID':[1, 2, 3,4],
                    'Gender':['Male','Male','Male','Female']})
ndf2=pd.DataFrame({'Employee ID':[1, 2, 3],
                  'Address':['North America','New York','Los angeles']})
display(ndf1)
display(ndf2)
ndf1.merge(ndf2[['Employee ID','Address']],on="Employee ID",how="left")
```

```python
#Q21. How do you select specific rows in a DataFrame using their indexes?
#Ans:rows based on integer indexing, we use .iloc function.
#rows based on label indexing,we use .loc function.
data2=pd.DataFrame({'Number1': [1,2,3,4],'Number2': [55,66,77,88]})
display(data2)
data2.iloc[[2]] #.iloc selects rows based on an integer index. So, if you want to select the 2nd row in a DataFrame, since the first row is at index 0, the second row is at index 1, and so on.
```

```python
data2.set_index("Number1", inplace=True) 
```

```python
data2.loc[[2]] #.loc selects rows based on a labeled index. So, if you want to select the row with an index label of 2, you would directly use df.loc[[2]].
```
